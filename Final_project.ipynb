# Final Project: Building a Terminal-based Naughts and Crosses Game (Tic Tac Toe)

**Description**

Your final project involves developing a fully functioning game of Naughts and Crosses (also known as Tic Tac Toe) that will be played directly in the terminal. This simple, two-player game will offer you a comprehensive way to demonstrate the programming skills you've learned throughout this course. 

**Objectives**

The main objectives of this project are as follows:

1. To design and implement a user-friendly terminal interface.
2. To demonstrate your understanding of the object-oriented programming (OOP) paradigm.
3. To create a game logic that determines the winner, loser, or a draw.
4. To manage user inputs and system outputs effectively.


**Requirements**

The game must have:

1. An interactive terminal interface for the user to interact with.
2. An option for the user to choose to be either 'X' or 'O'.
3. An easy-to-read game board on the terminal. 


4. The ability to take user inputs to determine where to place their 'X' or 'O'.
5. Clear indications of whose turn it is.
6. A system to check for three 'X's or 'O's in a row, either horizontally, vertically or diagonally.
7. A message declaring the end of the game, along with the winner or indicating a draw.
8. An option to restart the game after it ends.

**Project Steps**

Here's a high-level overview of the steps you might take to complete this project:

1. **Planning your project:** Write a brief outline of how you'll organize your project. This should include the classes and methods you plan to implement.

2. **Setting up your project:** Create a new Python project in your preferred development environment.

3. **Building the game board:** Start by creating the visual representation of the game board. This will be a 3x3 grid where the players can place their 'X's or 'O's.

4. **Taking player input:** Create a system for the player to place their symbol on the board. This could be as simple as asking for coordinates.

5. **Checking for a winner:** Implement the game logic. You should check the board after every move to see if there is a winner.

6. **Ending the game:** If there's a winner or the board is full, display a message to the players and give them the option to play again.

7. **Testing your game:** Make sure to test your game thoroughly. Try to break it in any way you can and fix any bugs that arise.

**Deliverables**

The final deliverables for this project will be:

1. A working version of your Naughts and Crosses game, with source code.

**Evaluation**

Your project will be evaluated based on:

1. The functionality of your game: Does it work as expected without any errors?
2. Your coding style: Is your code well-organized and easy to read, with clear comments explaining your logic?
3. Your problem-solving skills: How did you approach the design and implementation challenges, and how well did you overcome them?
